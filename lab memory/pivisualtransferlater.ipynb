```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('pi.txt',sep=' ')
df.head()
```

```python
%matplotlib inline
```

```python
df.plot(x="iteration", y=["leybnicpi"], color = 'c', lw = 0.1)
plt.legend( ['Leibnizs formula'])
plt.ylim(3.14,3.145)
plt.xlim(0, 10000)
plt.xlabel('Iteration')
plt.ylabel('Calculated value of pi')
plt.grid(c='#BFEFEF',ls='-',lw=0.5)
plt.title('Calculating Pi with Leibnizs formula')
```

```python
df.plot(x="iteration", y=["eylerpi"],color='#7F007F', lw = 1)
plt.legend( ['Eulers formula'])
plt.ylim(3.14,3.144)
plt.xlim(0, 10000)
plt.xlabel('Iteration')
plt.ylabel('Calculated value of pi')
plt.grid(c='#BFEFEF',ls='-',lw=0.5)
plt.title('Calculating Pi with Eulers formula')
```
